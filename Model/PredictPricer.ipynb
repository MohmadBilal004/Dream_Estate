```python
data1 = pd.read_csv("house_prices.csv")
data1.head()
```

```python
data1.shape
```

```python
data1['Brand New'] = data1['Title'].str.contains('Brand New').astype(int)
data1['Brand New'] = data1['Title'].apply(lambda x:1 if 'Brand New' in x else 0)
data1.head(50)
```

```python
# Replace the specified values in the 'Title' column with 'House'
data1
data1['Title'] = data1['Title'].apply(lambda x: 'House' if 'House' not in x else x)
# Show the first few rows of the updated dataframe
data1.head()
```

```python
# Modify the 'Title' column to keep only specific values
# specific_values = ['House', 'Luxury House', 'Apartment', 'Brand New', 'Villa', 'Hotel','Land']
specific_values = ['House', 'Luxury House', 'Apartment', 'Villa', 'Hotel']
data1['Title'] = data1['Title'].apply(lambda x: ' '.join([word for word in x.split() if word in specific_values]))
data1.groupby('Title')['Title'].agg('count')
data1['Title'].head()
```

```python
#Assumed only following columns are important
data2= data1.loc[:,['Title','Price','Baths','Land size', 'Beds','House size', 'Location', 'Brand New']]
```

```python
data2.head()
```

```python
data2.isnull().sum()
```

```python
data2.head()
```

```python
# data2['Location'].unique()
data2[['Town','District']]=data2.Location.str.split(",",expand=True)
data2= data2.drop('Location',axis=1)
```

```python
data2['Price'] = data2['Price'].str.replace('Rs', '').str.replace(',', '').astype(int)
data2['House size'] = data2['House size'].str.replace('sqft', '').str.replace(',', '').astype(float)
data2.head()
```

```python
def acres_to_perches(row):
    if(len(row['Land size'].split())!=2):
        quantity = 'NaN'
    else:
        quantity, uom = row['Land size'].split()
        quantity = quantity.replace(',', '')
        if uom == 'acres':
            quantity=quantity*43560
    return quantity

data2['Land size'] = data2.apply(acres_to_perches, axis=1)
```

```python
def is_int(x):
    try:
        int(x)
    except:
        return False
    return True

data2= data2[data2['Baths'].apply(is_int)]
data2= data2[data2['Beds'].apply(is_int)]
data2.shape
```

```python
data2['Baths'] = data2['Baths'].astype('int32')
data2['Beds'] = data2['Beds'].astype('int32')
data2.dtypes
```

```python
data2['Land size'] = pd.to_numeric(data2['Land size'], errors='coerce')

# Now convert to float
data2['Land size'] = data2['Land size'].astype(float)
data2.head()
```
